1. What is a Vector Database (VectorDB) and how is it different from
traditional databases?
         
     - A Vector Database (VectorDB) is a type of database designed to store, index, and search high-dimensional vector embeddings (numerical representations of data such as text, images, or audio generated by AI/ML models). It enables similarity search instead of exact keyword matching.

Vector databases are commonly used in AI applications like semantic search, recommendation systems, and Retrieval-Augmented Generation (RAG).

Difference between Vector Database and Traditional Database:

1. Data Type:
- VectorDB: Stores vectors/embeddings
- Traditional DB: Stores structured data (rows/columns)

2. Search Type:
- VectorDB: Similarity search (semantic search)
- Traditional DB: Exact match / keyword search

3. Query Method:
- VectorDB: Distance metrics (cosine, Euclidean)
- Traditional DB: SQL queries

4. Use Case:
- VectorDB: AI search, recommendations, NLP
- Traditional DB: Banking, inventory, CRUD systems

5. Structure:
- VectorDB: High-dimensional vectors
- Traditional DB: Tables (relational/NoSQL)

3. Why is Chroma DB considered important in the context of AI/ML projects?
Describe its key features.

   - Chroma DB is an open-source vector database that is widely used in AI/ML projects for storing and managing embeddings generated from text, images, or other data types. It is important because it enables efficient similarity search, which is a key requirement in modern AI applications like semantic search, chatbots, and Retrieval-Augmented Generation (RAG).

   Chroma DB is especially popular in LLM-based applications because it is simple to integrate with Python and works smoothly with frameworks like LangChain and LlamaIndex.

Key Features of Chroma DB:

1. Vector Storage:
Stores high-dimensional embeddings generated by AI models.

2. Similarity Search:
Allows fast retrieval of semantically similar data using vector distance metrics.

3. Easy Integration:
Works seamlessly with Python, LangChain, and LlamaIndex.

4. Persistent Storage:
Supports saving data locally so embeddings are not lost after session ends.

5. Metadata Support:
Allows storing additional information (like tags, documents, IDs) along with vectors.

6. Scalability:
Can handle small to medium-scale AI applications efficiently.

7. Developer-Friendly:
Simple API design makes it easy for beginners and fast prototyping.

Because of these features, Chroma DB is widely used in RAG pipelines, AI chatbots, and document search systems.

```python
# Install ChromaDB
!pip install chromadb

import chromadb
from chromadb.config import Settings

# Create client (in-memory)
client = chromadb.Client(Settings(anonymized_telemetry=False))

# Create collection (like a vector table)
collection = client.create_collection(name="my_vectors")

# Add sample documents with embeddings auto-handled internally
collection.add(
    documents=[
        "AI is transforming the world",
        "Chroma DB is used for vector search",
        "Python is popular for machine learning"
    ],
    ids=["1", "2", "3"]
)

# Query similar documents
results = collection.query(
    query_texts=["vector database in AI"],
    n_results=2
)

print(results)
```

📦 4. Simple Retrieval (Similarity Search using keywords)

```python
def retrieve(query, docs):
    query_words = set(query.lower().split())
    scored = []

    for doc in docs:
        doc_words = set(doc.lower().split())
        score = len(query_words.intersection(doc_words))
        scored.append((score, doc))

    scored.sort(reverse=True)
    return [doc for score, doc in scored[:2]]
```

📦 5. Call Llama3 via Ollama API

```python
import requests

def generate_response(context, query):
    url = "http://localhost:11434/api/generate"

    prompt = f"""
Use the context below to answer the question.

Context:
{context}

Question:
{query}
"""

    data = {
        "model": "llama3",
        "prompt": prompt,
        "stream": False
    }

    response = requests.post(url, json=data)
    return response.json()["response"]
```

    🧠 System Idea

     A health-tech chatbot can be built using a RAG (Retrieval-Augmented Generation) pipeline:

     Hugging Face Models (Understanding)
     Convert medical text into embeddings (sentence representations)
     Example model: sentence-transformers

📦 1. Install Libraries

```python
!pip install chromadb sentence-transformers requests -q
```

📦 2. Import Libraries

```python
import chromadb
from sentence_transformers import SentenceTransformer
import requests
```

📦 3. Load Embedding Model (Hugging Face)

```python
embedder = SentenceTransformer('all-MiniLM-L6-v2')
```

📦 4. Create VectorDB (ChromaDB)

📦 5. Sample Medical Articles

```python
articles = [
    "Diabetes is a chronic disease affecting blood sugar levels.",
    "Hypertension increases risk of heart diseases and stroke.",
    "COVID-19 spreads through respiratory droplets.",
    "Regular exercise improves cardiovascular health."
]

embeddings = embedder.encode(articles).tolist()

collection.add(
    documents=articles,
    embeddings=embeddings,
    ids=["1", "2", "3", "4"]
)
```

📦 6. Retrieval Function (Vector Search)

```python
def retrieve(query):
    query_embedding = embedder.encode([query]).tolist()

    results = collection.query(
        query_embeddings=query_embedding,
        n_results=2
    )

    return results["documents"][0]
```

📦 7. Ollama (Llama3 Generation)

```python
def generate_answer(context, query):
    url = "http://localhost:11434/api/generate"

    prompt = f"""
You are a medical assistant chatbot.

Context:
{context}

Question:
{query}

Answer in simple language:
"""

    response = requests.post(url, json={
        "model": "llama3",
        "prompt": prompt,
        "stream": False
    })

    return response.json()["response"]
```
